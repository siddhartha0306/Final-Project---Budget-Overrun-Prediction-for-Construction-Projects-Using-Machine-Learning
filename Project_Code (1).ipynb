```python
df = df[(df["totalEstimatedCost"] > 0) & (df["TotalIncurredCost"] >= 0)]

df["budget_overrun"] = (df["TotalIncurredCost"] > df["totalEstimatedCost"]).astype(int)

df["budget_overrun"].value_counts()
```

```python
df_model = df.drop(columns=["totalEstimatedCost", "TotalIncurredCost"])
```

```python
y = df_model["budget_overrun"]

# Identify and drop columns that are descriptive strings or dates and not useful for modeling
# Also drop 'Unnamed: 29' as it causes warnings due to being entirely empty.
drop_for_model = [
    'projectRegId', 'projectName', 'promoterName', 'architect_name', 'eng_name',
    'projectAddress', 'projectAddress2', 'startDate', 'completionDate',
    'tPNo', 'tpo_code', 'location_coordinates', 'Unnamed: 29' # Added 'Unnamed: 29'
]
X = df_model.drop(columns=drop_for_model + ["budget_overrun"], errors="ignore")
```

```python
categorical_cols = ["projectType", "promoterType", "underRedevelopment", "distName"]
numeric_cols = [c for c in X.columns if c not in categorical_cols]
```

```python

# Automatically detect categorical (object, category) and numeric (int, float) columns
categorical_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
numeric_cols = X.select_dtypes(include=["number"]).columns.tolist()

print("Categorical columns:", categorical_cols)
print("Numeric columns:", numeric_cols)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Filter out 'Unnamed: 29' from numeric_cols to prevent the warning
filtered_numeric_cols = [col for col in numeric_cols if col != 'Unnamed: 29']

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, filtered_numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

def evaluate_model(name, y_true, y_pred):
    print(f"\n{name}")
    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall   :", recall_score(y_true, y_pred))
    print("F1 Score :", f1_score(y_true, y_pred))
```

# MODELS

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", KNeighborsClassifier(
        n_neighbors=5,        # you can tune this
        weights="distance"    # closer neighbours have more influence
    ))
])

knn_clf.fit(X_train, y_train)
pred_knn = knn_clf.predict(X_test)

evaluate_model("K-Nearest Neighbours", y_test, pred_knn)
```

```python
# KNN model
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

plt.figure(figsize=(5,4))
ConfusionMatrixDisplay.from_predictions(y_test, pred_knn, cmap="Blues")
plt.title("Confusion Matrix - KNN")
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier

nn_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", MLPClassifier(
        hidden_layer_sizes=(128, 64),
        activation="relu",
        solver="adam",
        max_iter=500,
        random_state=42
    ))
])

nn_clf.fit(X_train, y_train)
pred_nn = nn_clf.predict(X_test)

evaluate_model("Neural Network (MLP)", y_test, pred_nn)
```

```python
plt.figure(figsize=(5,4))
ConfusionMatrixDisplay.from_predictions(y_test, pred_nn, cmap="Blues")
plt.title("Confusion Matrix - Neural Network")
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", DecisionTreeClassifier(
        max_depth=6,        # you can tune this
        random_state=42
    ))
])

dt_clf.fit(X_train, y_train)
pred_dt = dt_clf.predict(X_test)

evaluate_model("Decision Tree", y_test, pred_dt)
```

```python
plt.figure(figsize=(5,4))
ConfusionMatrixDisplay.from_predictions(y_test, pred_dt, cmap="Blues")
plt.title("Confusion Matrix - Decision Tree")
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("clf", GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.1,
        max_depth=3,
        random_state=42
    ))
])

gb_clf.fit(X_train, y_train)
pred_gb = gb_clf.predict(X_test)

evaluate_model("Gradient Boosting Classifier", y_test, pred_gb)
```

```python
plt.figure(figsize=(5,4))
ConfusionMatrixDisplay.from_predictions(y_test, pred_gb, cmap="Blues")
plt.title("Confusion Matrix - Gradient Boosting Classifier")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

results = pd.DataFrame({
    "Model": [ "K-Nearest Neighbours", "Decision Tree", "Neural Network","Gradient Boosting Classifier"],
    "Accuracy": [
        accuracy_score(y_test, pred_knn),
        accuracy_score(y_test, pred_dt),
        accuracy_score(y_test, pred_nn),
        accuracy_score(y_test, pred_gb)
    ],
    "Precision": [

        precision_score(y_test, pred_knn),
        precision_score(y_test, pred_dt),
        precision_score(y_test, pred_nn),
        precision_score(y_test, pred_gb)
    ],
    "Recall": [

        recall_score(y_test, pred_knn),
        recall_score(y_test, pred_dt),
        recall_score(y_test, pred_nn),
        recall_score(y_test, pred_gb)
    ],
    "F1 Score": [

        f1_score(y_test, pred_knn),
        f1_score(y_test, pred_dt),
        f1_score(y_test, pred_nn),
        f1_score(y_test, pred_gb)
    ]
})

print(results)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(6, 5))


proba_knn = knn_clf.predict_proba(X_test)[:, 1]
proba_dt = dt_clf.predict_proba(X_test)[:, 1]
proba_nn = nn_clf.predict_proba(X_test)[:, 1]
proba_gb = gb_clf.predict_proba(X_test)[:, 1]

#  for KNN
fpr_knn, tpr_knn, _ = roc_curve(y_test, proba_knn)
auc_knn = auc(fpr_knn, tpr_knn)
plt.plot(fpr_knn, tpr_knn, label=f"KNN (AUC = {auc_knn:.3f})")

# Decision Tree
fpr_dt, tpr_dt, _ = roc_curve(y_test, proba_dt)
auc_dt = auc(fpr_dt, tpr_dt)
plt.plot(fpr_dt, tpr_dt, label=f"Decision Tree (AUC = {auc_dt:.3f})")

# Neural Network
fpr_nn, tpr_nn, _ = roc_curve(y_test, proba_nn)
auc_nn = auc(fpr_nn, tpr_nn)
plt.plot(fpr_nn, tpr_nn, label=f"Neural Net (AUC = {auc_nn:.3f})")

# Gradient Boosting
fpr_gb, tpr_gb, _ = roc_curve(y_test, proba_gb)
auc_gb = auc(fpr_gb, tpr_gb)
plt.plot(fpr_gb, tpr_gb, label=f"Gradient Boosting (AUC = {auc_gb:.3f})")

# Diagonal
plt.plot([0, 1], [0, 1], linestyle="--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison - All Models")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```
